Imbalanced Dataset:
A dataset with an unequal class distribution:

```python
#importing the dependencies
import numpy as np
import pandas as pd
#loading the dataset to pandas dataFrame
credit_card_data=pd.read_csv('/content/credit_data.csv')
#First 5 rows of the data Frame
credit_card_data.head()
```

```python
 credit_card_data.tail()
```

```python
#distibution of the two classes
credit_card_data['Class'].value_counts()
```

This is a highly imbalanced data set


```python
#Separating the legit and fraudulent transactions
```

```python
legit= credit_card_data[credit_card_data['Class']==0]
fraud= credit_card_data[credit_card_data['Class']==1]
print(legit.shape)
print(fraud.shape)
```

Under-Sampling

Build a sample datset containing similar distribution of legit and fraud transaction

```python
legit_sample=legit.sample(n=365)
```

```python
legit_sample.shape
```

Concatanate the two Data Frames

```python
new_dataset= pd.concat([legit_sample,fraud],axis=0)
```
